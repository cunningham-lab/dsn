```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import matplotlib

fontsize = 26
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : fontsize}

matplotlib.rc('font', **font)
```

```python
def plot_SC_errs(err1, err2, color, fname):
    val1 = 1
    val2 = 2
    barwidth = 0.2
    plt.figure()
    plt.plot([val1-barwidth, val1+barwidth], [err1, err1], c=color, lw=8)
    plt.plot([val2-barwidth, val2+barwidth], [err2, err2], c=color, lw=8)
    plt.xlim([0, 3])
    plt.ylim([0, 0.5])
    
    ax = plt.gca()
    
    ax.set_xticks([1, 2])
    ax.set_xticklabels(['Pro', 'Anti'], fontsize=fontsize)
    
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_ylabel('error increase', weight='bold', fontsize=fontsize)
    plt.tight_layout()
    plt.savefig(fname)
    plt.show()
    

fname = '/Users/sbittner/Documents/dsn/written/Journal2019/figs/fig1/SC_model_behavior.png'
plot_SC_errs(0.05, 0.2, np.array([51, 151, 51])/256.0, fname)

fname = '/Users/sbittner/Documents/dsn/written/Journal2019/figs/fig1/SC_model_random.png'
plot_SC_errs(0.1, 0.02, np.array([153, 51, 51])/256.0, fname)
```
