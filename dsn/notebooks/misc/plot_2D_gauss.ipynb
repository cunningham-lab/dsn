```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import multivariate_normal as mvn

```

```python
mu = np.zeros((2,))
Sigma = np.eye(2)
dist = mvn(mu, Sigma)
```

```python
x = np.random.normal(0.0, 1.0, (2, 1000))
logpdf = dist.logpdf(x.T)
```

```python
plt.figure(figsize=(4,4))
plt.plot([0,0], [-4, 4], 'k--', lw=1)
plt.plot([-4, 4], [0, 0], 'k--', lw=1)
plt.scatter(x[0,:], x[1,:], s=5, c=logpdf)
plt.xlim([-4, 4])
plt.ylim([-4, 4])
plt.axis('off')
plt.show()
```
