```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

T = [10, 20, 200, 500]

#MB
aug_lag_memory = [64.2, 130, 674]

# ms
aug_lag_time = [43.4, 77.5, 402]
```

```python
num_T = 3
fontsize = 14
labels = ['Aug Lag Cost (grads)', 'tensorboard dies']
x = np.log(T[:num_T])
plt.figure()
plt.plot(x, aug_lag_time[:num_T], 'g')
plt.plot(np.log(T[-1]), [0], c='k', marker='*')
ax = plt.gca()
ax.set_xticks(np.log(T))
ax.set_xticklabels(np.array(T))
plt.xlabel('T')
plt.ylabel('operation time (ms)', fontsize=fontsize)
plt.legend(labels, fontsize=fontsize)
plt.show()
```

```python
num_T = 3
fontsize = 14
labels = ['Aug Lag Cost (grads)']
x = np.log(T[:num_T])
plt.figure()
plt.plot(x, aug_lag_memory[:num_T], 'g')
plt.scatter(np.log(T[-1]), [0], c='k', marker='*')
ax = plt.gca()
ax.set_xticks(np.log(T))
ax.set_xticklabels(np.array(T))
plt.xlabel('T')
plt.ylabel('operation memory (Mb)', fontsize=fontsize)
plt.legend(labels, fontsize=fontsize)
plt.show()
```
